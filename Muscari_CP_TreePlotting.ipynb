```python
## import
import ipyrad as ip
import ipyrad.analysis as ipa
import ipyparallel as ipp
import pandas as pd
import toytree
import toyplot

## print Version of ipyrad und toytree
print("ipyrad v. {}".format(ip.__version__))
print("toytree v. {}".format(toytree.__version__))

## print Version of Python
from platform import python_version
print("Python v.", python_version())
```

## Plotting MrBayes Tree without indel codding

```python
## assign tree and location to variable
mbtree = "/home/tim/Phylo/Muscari_CP/Mus_CP_MrBayes_20210811/Muscari_CP_HSEX_INVREF_MrBayes_cons.tre"

## load from the .trees attribute or from the saved tree file
tre = toytree.tree(mbtree, tree_format=10).root(wildcard = "Brimeura")

## draw the tree with styling
canvas, axes, mark = tre.ladderize(1).draw(
    width = 900,
    height = 900,
    tip_labels_align=True,
    scalebar=True,
    tip_labels_style = {"font-size": "11px"},
    node_labels='prob',
    node_sizes=0,
    node_labels_style={"font-size": "12px",
                       "baseline-shift": "7px",
                       "-toyplot-anchor-shift": "-13px"},
);
```

```python
import toyplot.pdf
toyplot.pdf.render(canvas, "/home/tim/Phylo/Muscari_CP/Mus_CP_MrBayes_20210811/Muscari_CP_MrBayes_20210811.pdf");
```
